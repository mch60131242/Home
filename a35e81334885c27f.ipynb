```python
x_train = train_data['document']
y_train = train_data['label']

x_test = test_data['document']
y_test = test_data['label']
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
## 딥러닝을 활용한 자연어 처리
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Embedding
from tensorflow.keras.utils import to_categorical
```

```python
#  자연어 전처리 -> 토큰화
token = Tokenizer()
token.fit_on_texts(x_train)  
list(token.word_index.keys())[:10]
```

```python
x_data = {}
for key, value in token.word_index.items():
    if value > 2:
        x_data[key] = values

print(len(x_data))
x = list(x_data.keys())
x
```

```python
# 서로 다른 길이의 데이터를 padding -> 4로 
padd_x = token.texts_to_sequences(x_train)   # 단어를 토큰화 하고 인덱스만 가져옴

# 서로 다른 길이의 데이터를 padding -> 4로 
padded_x = pad_sequences(padd_x, 10)
print(" 패딩한 결과 출력 : ", padded_x)
```

```python
# 임베딩에 입력된 단어의 수를 지정
word_size = len(token.word_index) + 1

# 단어 임베딩을 포함한 딥러닝 모델 생성
model = Sequential()
model.add(Embedding(word_size, 8, input_length=10))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
#  model 실행 옵션 설정
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(padded_x, y_train, epochs=20)
```

```python
# 서로 다른 길이의 데이터를 padding -> 4로 
padd_xt = token.texts_to_sequences(x_test)   # 단어를 토큰화 하고 인덱스만 가져옴

# 서로 다른 길이의 데이터를 padding -> 4로 
padded_xt = pad_sequences(padd_xt, 10)
model.evaluate(padded_xt, y_test)
```
