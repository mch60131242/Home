```python
import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('./fig/puppy.bmp', 1)
glass = cv2.imread('./fig/imgbin_sunglasses_1.png', cv2.IMREAD_UNCHANGED)
glass = cv2.resize(glass, (300, 150))

mask = glass[:, :, -1]
glass_ = glass[:, :, :-1]

h, w = mask.shape[:2]

src_crop = src[120:120+h, 220:220+w]
cv2.copyTo(glass_, mask, src_crop)

if src is None or glass is None:
    print('images read failed')
    sys.exit()
    
cv2.imshow('src', src)
cv2.imshow('src_crop', src_crop)
cv2.imshow('glass_', glass_)
cv2.imshow('mask', mask)


cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import glob

img_paths = glob.glob('../Ch_1_Opencv_Intro/fig/my_images/*.*')

# display(img_paths)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN,
                      cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(img_paths[idx])
    
    if img is None:
        print('image read failed')
        break
    
    cv2.imshow('image', img)
    
    if cv2.waitKey(3000) == 27:
        break
    
    idx += 1
    
    if idx >= len(img_paths):
        idx = 0
        
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./fig/green.png')

cv2.imshow('image', img)

while True:
    key = cv2.waitKey()
    
    if key == 27 or key == ord('q'):
        break
    
    elif key == ord('e'):
        img = cv2.Canny(img, 100, 150)
        cv2.imshow('image', img)
        
cv2.destroyAllWindows()
```

```python
img = np.ones((600, 1200, 3), np.uint8)*255

# cv2.line(img, pt1, pt2, color, thickness = None, lineType = None, shift = None) -> img
# flags
    # img:그림을 그릴 영상
    # pt1, pt2: 직선의 시작점, 끝점
    # color: 직선의 칼라 (B,G,R)의 튜플
    # thinkness: 선두께, 기본은= 1
    # lineType: cv2.LINE_4, cv2.LINE_8, cv2.LINE_AA
    
cv2.line(img, (100, 50), (300, 50), (0, 0, 255), 10, cv2.LINE_8)
cv2.line(img, (300, 50), (200, 300), (0, 0, 255), 10, cv2.LINE_AA)
cv2.arrowedLine(img, (400, 50), (400, 300), (255, 0, 0), 10, cv2.LINE_AA)

# cv2.rectangle(img, pt1, pt2, color, thickness = None, lineType = None) -> img
   # pt1 :좌측 상단,  pt2: 우측하단
# cv2.rectangle(img, rect, color, thickness = None, lineType = None) -> img
    # rect: 사각형의 위치 정보 (x, y, w, h)
cv2.rectangle(img, (100, 400), (400, 500), (0, 0, 255), 5)
cv2.rectangle(img, (120, 420, 260, 60), (0, 255, 0), -1)

# cv2.circle(img, center, radius, color, thickness = None, lineType = None) -> img
    # center: 원의 중심좌표 (x, y)
    # radius : 원의 반지름
cv2.circle(img, (800, 300), 100, (43, 167, 200), 10, cv2.LINE_AA)

# cv2.ellips(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]]) -> img
    # center: 원의 중심좌표 (x, y)
    # axis: 축의 반지름(x, y)
    # angle: 타원의 기울기 (예, 10, 오른쪽으로 10도 기울어짐)
    # startAngle: 타원을 그리는 시작 각도 (3시 방향이 0도)
    # endAngle: 타원을 그리는 종료 각도
    
cv2.ellipse(img, (800, 300), (50, 200), -30, 0, 360, (0, 0, 255), 10, cv2.LINE_AA)

##
cv2.putText(img, 'Hello opencv!', (800, 500), cv2.FONT_HERSHEY_SIMPLEX,
            1, (255,  0, 0), 1, cv2.LINE_AA)


cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## 
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('camera open failed')
    sys.exit()

w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# print('width = ', w)
# print('height = ', h)
# print('fps = ', fps)
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('output.avi', fourcc, fps, (w, h))
    
while True:
    
    ret, frame = cap.read()
    
    if not ret:
        print('frame read failed')
        break
    
    edge = cv2.Canny(frame, 50, 100)
    print('edge shape = ',edge.shape)
    
    out.write(frame)
    
    cv2.imshow('frame', frame)
    cv2.imshow('edge', edge)
    
    if cv2.waitKey(20) == 27:
        break
    
out.release()
cap.release()
cv2.destroyAllWindows()
    
```
