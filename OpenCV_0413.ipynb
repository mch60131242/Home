```python
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print('1')
    sys.exit()
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))

h =   int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fps = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'DVIX')
out = cv2.VideoWriter('kkk.avi',fourcc,fps,(w,h))
while True:
    ret,frame =cap.read()
    if not ret:
        print('1')
        break
    out.write(frame)
    cv2.imshow('1',frame)
    
    
    if cv2.waitKey(20)==27:
        break
out.release()
cap.release()
cv2.destroyAllWindows()
```

```python
## 키보드방식으로 입력받아 만들기 인버젼,
```

```python
img =cv2.imread('./test/fig/puppy.bmp',0)
img_ori =img.copy()
if img is None:
    print('1')
    sys.exit()
cv2.imshow('1',img)
while True:
    key = cv2.waitKey()
    if key == 27 or key == ord('q'):
        break
        
    elif key == ord('e'):
        img =cv2.Canny(img, 50,100)
        cv2.imshow('1',img)
    elif key == ord('i'):
        img = 255 - img
        cv2.imshow('1',img)
    
    elif key ==ord('r'):
        img = img_ori.copy()
        cv2.imshow('1',img_ori)
    elif key ==ord('s'):
        cv2.imwrite('this.png',img)

cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print('1')
    sys.exit()
f_flags = False ## 키를 외부에 설정하여 한번만 눌러도 반복문안에서 지속적으로 사용되게 하는법ㅠ
e_flags = False

while True:
    ret,frame =cap.read()
    
    if f_flags:
        frame = cv2.Canny(frame,50,100)
    
    if e_flags :
        frame = cv2.flip(frame,1)
    cv2.imshow('1',frame)
    
    key =cv2.waitKey(20)
    
    if key == 27:
        break
    elif key ==ord('f'):
        f_flags = not f_flags
        
    elif key ==ord('e'):
        e_flags = not e_flags
        
        
    

        
    

cap.release()
cv2.destroyAllWindows()
```

```python
img =np.ones((600,1200,3),np.uint8)*255
img1=img.copy()






def call_mouse(event,x,y, flags,param):
    global oldx,oldy
    
    
    if event ==cv2.EVENT_LBUTTONDOWN:
        oldx,oldy = x,y
        
    elif event ==cv2.EVENT_MOUSEMOVE:
        if flags ==cv2.EVENT_FLAG_LBUTTON:
            cv2.line(img,(oldx,oldy), (x,y),(255,0,255),10,cv2.LINE_AA)
            cv2.imshow('1',img)
            
            oldx,oldy =x,y
cv2.namedWindow('1')            
cv2.setMouseCallback('1',call_mouse)
cv2.imshow('1',img)


while True:
    key = cv2.waitKey()
    
    if key == 27:
        break
    elif key ==ord('s'):
        cv2.imwrite('this.png',img)
    elif key ==ord('r'):
        img = img1.copy()
        cv2.imshow('1',img)
        

cv2.destroyAllWindows()
```
