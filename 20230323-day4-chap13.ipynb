```python
# 광물 데이터 로드
import pandas as pd

df = pd.read_csv('./data/sonar3.csv', header=None)
df.head()
```

```python
df[60].value_counts()   # 광석과 일반광석 으로 분류
# df.info()

X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
# 초음파 광물 예측하기 : 모델 설정 및 실행
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#  모델 설정
model = Sequential()
model.add(Dense(24, input_dim=X.shape[1], activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, batch_size=10)
```

```python
# 광물 데이터를 학습데이터와 테스트 데이터로 분리
from sklearn.model_selection import train_test_split

#  학습셋과 테스트셋으로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=42)
```

```python
print(len(X_train), len(X_test))
print(X_train.shape, y_train.shape)
```

```python
# 모델 설정 
model = Sequential()
model = Sequential()
model.add(Dense(10, input_dim=X.shape[1], activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#   학습 데이터로 모델 실행 
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=100, batch_size=10)
```

```python
# 테스트 데이터로 모델 정확도 구함
score = model.evaluate(X_test, y_test)
print("test score : ", score[1])
```

```python
def model_make(input_dim):   #   모델 구조 설계
    model = Sequential()   # 모델 초기화, 시작
    model.add(Dense(24, input_dim=input_dim, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))    
    return model
 
def model_compile(model, X, y,loss, epochs):   # 모델 실행
    model.compile(loss=loss, optimizer='adam', metrics=['accuracy'])
    return model.fit(X, y, epochs=epochs, verbose=False)

def model_validation(model, X, y):
    return model.evaluate(X, y)[1]
```

```python
df = pd.read_csv('./data/sonar3.csv', header=None)
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
# 5. k겹 교차 검증
from sklearn.model_selection import KFold

k = 5  # 몇겹? -> 5겹
kfold = KFold(n_splits=k, shuffle=True)

acc_score = []
models = []
for train_index, test_index in kfold.split(X):
#     print("train : ", train_index, "\ntest : ",test_index)
    X_train, X_test = X.iloc[train_index, : ], X.iloc[test_index, :]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model = model_make(X_train.shape[1])  # X_train, y_train 데이터를 가지고 모델 학습  -> 함수로 정의
    history = model_compile(model, X_train, y_train, 'binary_crossentropy', 100)
    acc_score.append(model_validation(model, X_test, y_test)) # X_test, y_test를 가지고 검증한 결과를 acc_score 리스트에 추가
    models.append(model)

# 결과 출력
print(' 정확도 : ', acc_score)
print(' 정확도의 평균 : ', sum(acc_score)/k)
```
