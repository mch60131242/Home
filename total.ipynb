```python
 
    
    def cal():
        
        traditional_input = []
        trend_input = []
        global traditional_market
        global trend_market
        
        
        information_data=int(input(('시장과 대형마트중 골라주세요!(1번은 시장,2번은 대형마트)')))
       
        if information_data ==1:
            while True:
                input_data = input("원하는 물품을 입력하세요,종료는 q입니다")

                if input_data == 'q':
                    break

                elif input_data  in list(traditional_market['품목 이름'].value_counts().index):
                    print('입력받았습니다.')
                    traditional_input.append(input_data)

                elif  input_data  not in list(traditional_market['품목 이름'].value_counts().index) :
                    print('없는 물품입니다!')
            traditional_sum=0
            trend_sum = 0
            for i in traditional_input:
                
                if i in list(traditional_market['품목 이름'].value_counts().index):
                    traditional_sum = traditional_sum+(round(traditional_market[traditional_market['품목 이름']==i]['가격(원)'].mean()))
            print('시장에서 주문하신 {}의 총액은 {}입니다'.format(traditional_input,traditional_sum))
            for i in traditional_input:
                if i in list(trend_market['품목 이름'].value_counts().index):
                    trend_sum = trend_sum+(round(trend_market[trend_market['품목 이름']==i]['가격(원)'].mean()))
            print('위에 입력하신 품목을 마트에서 구매하실 경우 금액은',trend_sum,'이며','두금액의 차는',traditional_sum-trend_sum,'입니다')
                
            
        elif information_data ==2:
            
            while True:
                input_data = input("원하는 물품을 입력하세요,종료는 q입니다")

                if input_data == 'q':
                    break
 
                elif input_data  in list(trend_market['품목 이름'].value_counts().index):
                    print('입력받았습니다.')
                    trend_input.append(input_data)

                elif  input_data  not in list(trend_market['품목 이름'].value_counts().index) :
                    print('없는 물품입니다!')
            trend_sum=0 
            traditional_sum=0
            for i in trend_input:
                
                if i in list(trend_market['품목 이름'].value_counts().index):
                    trend_sum = trend_sum+(round(trend_market[trend_market['품목 이름']==i]['가격(원)'].mean()))
            print('마트에서 주문하신 {}의 총액은 {}입니다'.format(trend_input,trend_sum))
            for i in trend_input:
                if i in list(traditional_market['품목 이름'].value_counts().index):
                    traditional_sum = traditional_sum+(round(traditional_market[traditional_market['품목 이름']==i]['가격(원)'].mean()))
            print('위에 입력하신 품목을 시장에서 구매하실 경우 금액은',traditional_sum,'이며','두금액의 차는',trend_sum-traditional_sum,'입니다')
            
```
