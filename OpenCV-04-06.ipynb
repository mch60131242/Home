```python
import os
```

```python
img_lists =os.listdir('./Ch_1_fig/fig/my_images/')
```

```python
img_lists
```

```python
## 밑의 문장은 너무 많이쓰는 거니까 외울것.
```

```python
img_paths = []

for i in img_lists:
    img_path = './Ch_1_fig/fig/my_images/' + i
    img_paths.append(img_path)

display(img_paths)
```

```python


cv2.namedWindow('1',cv2.WINDOW_NORMAL)
cv2.setWindowProperty('1',cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)



idx = 0
while True:
    img = cv2.imread(img_paths[idx])
    
    if img is None:
        print('없음')
        continue
        
    cv2.imshow('1',img)
    
    if cv2.waitKey(1000)==27:
        break
    
    idx += 1
    
    if idx>= len(img_paths):
        idx=0
        
        
        
cv2.destroyAllWindows()
```
